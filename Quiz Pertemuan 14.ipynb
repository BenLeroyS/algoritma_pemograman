```python
import numpy as np
from scipy import stats

data = {
    "Nama Siswa": ["Ali", "Budi", "Citra", "Dani", "Elina"],
    "Matematika": [85, 90, 88, 92, 87],
    "Fisika": [87, 79, 84, 77, 95],
    "Biologi": [75, 84, 86, 77, 87],
}


def hitung_statistik(nilai):
    mean = np.mean(nilai)
    median = np.median(nilai)
    mode = stats.mode(nilai)
    variance = np.var(nilai, ddof=1)  # Variasi sampel (ddof=1)
    std_deviation = np.std(nilai, ddof=1)
    return mean, median, mode, variance, std_deviation


for pelajaran in ["Matematika", "Fisika", "Biologi"]:
    nilai = data[pelajaran]
    mean, median, mode, variance, std_deviation = hitung_statistik(nilai)
    print(f"Statistik untuk {pelajaran}:")
    print(f"  Rata-rata: {mean}")
    print(f"  Median: {median}")
    print(f"  Mode: {mode}")
    print(f"  Variance: {variance}")
    print(f"  Standar Deviation: {std_deviation}\n")


print("Statistik per siswa:")
for i, nama in enumerate(data["Nama Siswa"]):
    nilai_siswa = [data["Matematika"][i], data["Fisika"][i], data["Biologi"][i]]
    mean, median, mode, variasi, std_deviation = hitung_statistik(nilai_siswa)
    print(f"{nama}:")
    print(f"  Rata-rata: {mean}")
    print(f"  Median: {median}")
    print(f"  Modus: {mode}")
    print(f"  Variance: {variance}")
    print(f"  Standar Deviation: {std_deviation}\n")
```
